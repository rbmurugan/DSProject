```python
#Import library
import sympy as sy
import numpy as np
from sympy.functions import cos
import matplotlib.pyplot as plt

plt.style.use("ggplot")

#Define variable and function
x = sy.Symbol('x')
f = cos(x)

# Function for Factorial

def factorial(n):
    if n <= 0:
        return 1
    else:
        return n*factorial(n-1)

# Taylor approximation at x0 of the function 'function'
def taylor(function,x0,n):
    i = 0
    p = 0
    while i <= n:
        p = p + (function.diff(x,i).subs(x,x0))/(factorial(i))*(x-x0)**i
        i += 1
    return p
```

```python
# Plot results
plt.figure(figsize=(8,8))
def plot():
    x_lims = [-6,6]
    x1 = np.linspace(x_lims[0],x_lims[1],800)
    y1 = []
    
    for j in range(0,4,1): #Modify here for the number of iterations
        func = taylor(f,0,j)  #Middle is the value at which x is considered, in this case x = 0
        print('Maclaurin approximation of order n='+str(j),func)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1,y1,label='order '+str(j))
        y1 = []
    # Plot the function to approximate (cosine in this case)
    plt.plot(x1,np.cos(x1),label='cos of x')
    plt.xlim(x_lims)
    plt.ylim([-5,5])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Maclaurin series approximation for cos')
    plt.show()

plot()
```
